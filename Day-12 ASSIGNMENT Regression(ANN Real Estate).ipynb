```python
x=data.iloc[:,1:7].values
y=data.iloc[:,7].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
ann=Sequential()
```

```python
ann.add(Dense(6,kernel_initializer="random_uniform",activation="relu"))
```

```python
ann.add(Dense(12,kernel_initializer="random_uniform",activation="relu"))
```

```python
ann.add(Dense(1,kernel_initializer="random_uniform",activation="relu"))
```

```python
ann.compile("adam","mse",metrics=["mse"])
```

```python
ann.fit(x_train,y_train,batch_size=32,epochs=1000)
```

```python
ypred=ann.predict(x_test)
```
